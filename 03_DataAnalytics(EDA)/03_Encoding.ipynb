```python
import pandas as pd
df = pd.read_csv(r"C:\Users\C9IN\Desktop\Mental_Health\02_Dataset\02_Clean_dataset.csv")
```

```python
unique_dict = {}

for col in df.columns:
    if col in ["Age", "Income"]:
        unique_dict[col] = {
            "min": df[col].min(),
            "max": df[col].max()
        }
    else:
        unique_dict[col] = df[col].unique().tolist()

unique_dict
```

1.Label Encoding (Ordered)

Alcohol_Consumption → Low< Moderate < High 

Sleep_Patterns → Poor < Fair < Good	

Physical_Activity_Level →	Sedentary < Moderate < Active	

Dietary_Habits	Unhealthy  → < Moderate < Healthy	

Education → High School < Associate < Bachelor < Master < PhD

1.a)Convert Alcohol Consumption

```python
df["Alcohol_Consumption"] = df["Alcohol_Consumption"].map({
"Low":0,
"Moderate":1,
"High":2
})
```

1.b)Convert Sleep Patterns

```python
df["Sleep_Patterns"] = df["Sleep_Patterns"].map({
"Poor":0,
"Fair":1,
"Good":2
})
```

1.c)Convert Physical Activity

```python
df["Physical_Activity_Level"] = df["Physical_Activity_Level"].map({
"Sedentary":0,
"Moderate":1,
"Active":2
})
```

1.d)Convert Dietary Habits

```python
df["Dietary_Habits"] = df["Dietary_Habits"].map({
"Unhealthy":0,
"Moderate":1,
"Healthy":2
})
```

1.e)Convert Education Level

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Education_Level"] = le.fit_transform(df["Education_Level"])
```

2. Binary Encoding

Fix Yes / No Values

```python
cols = [
    "History_of_Mental_Illness",
    "History_of_Substance_Abuse",
    "Family_History_of_Depression",
    "Chronic_Medical_Conditions"
]

df[cols] = df[cols].replace({"Yes":1, "No":0, "Y":1, "N":0}).astype(int)
```

Convert Employment Status

```python

df["Employment_Status"] = df["Employment_Status"].map({
"Unemployed":0,
"Employed":1
})
```

3.One-Hot Encoding

Fix Marital_Status 

```python
mapping = {
    'S': 'Single',
    'M': 'Married',
    'D': 'Divorced',
    'W': 'Widowed',
}

df["Marital_Status"] = df["Marital_Status"].replace(mapping)
```

```python
df = pd.get_dummies(df, columns=["Marital_Status"], dtype=int)
```

Convert Smoking Status

```python
df = pd.get_dummies(df, columns=["Smoking_Status"], dtype=int)
```
