6.GroupBy Analysis

GroupBy analysis is used to analyze the relationship between different features by grouping the data. It helps in identifying patterns and trends across categories.

```python
# Average Age by Smoking Status
df.groupby('Smoking Status')['Age'].mean()
```

```python
# Average Age by Marital Status
df.groupby('Marital Status')['Age'].mean()
```

```python
# Income distribution by Employment Status (count instead of mean)
df.groupby('Employment Status')['Income'].count()
```

```python
# Average number of children by Marital Status
df.groupby('Marital Status')['Number of Children'].mean()
```

7. Correlation (Numeric Columns)

Correlation analysis is used to measure the relationship between numerical features in the dataset. It helps in identifying how strongly variables are related to each other.
 

```python
print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

8. Pivot_table

Pivot table analysis is used to summarize data by aggregating values across multiple dimensions. It helps in gaining deeper insights by combining multiple features.

```python
# Average age based on smoking and marital status
df.pivot_table(values='Age', index='Smoking Status', columns='Marital Status', aggfunc='mean')
```

```python
# Count of records based on employment and smoking status
df.pivot_table(values='Income', index='Employment Status', columns='Smoking Status', aggfunc='count')
```

```python
# Average number of children by marital status and employment
df.pivot_table(values='Number of Children', index='Marital Status', columns='Employment Status', aggfunc='mean')
```

9. Skewness

Skewness and kurtosis are used to understand the distribution of numerical features. Skewness measures the asymmetry of data, while kurtosis indicates the presence of outliers.

```python
df[['Age', 'Number of Children']].agg(['skew', 'kurtosis'])
```

Age :

  Skew = 1.80 → Positively skewed distribution

  Kurt = 17.38 → Presence of outliers

Number of Children :

  Skew = 14.91 → Highly positively skewed

  Kurt = 291.81 → Significant presence of extreme outliers

```python
import matplotlib.pyplot as plt
df['Age'].plot(kind='kde')
plt.title("Age Density Curve")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
