```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv(r"C:\Users\C9IN\Desktop\Mental_Health\02_Dataset\04_Final_dataset.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_viz = df.copy()

df_viz['Sleep_Patterns'] = df_viz['Sleep_Patterns'].map({0:'Poor',1:'Fair',2:'Good'})
df_viz['Physical_Activity_Level'] = df_viz['Physical_Activity_Level'].map({0:'Sedentary',1:'Moderate',2:'Active'})
df_viz['Alcohol_Consumption'] = df_viz['Alcohol_Consumption'].map({0:'Low',1:'Moderate',2:'High'})
df_viz['Mental_Health_Level'] = df_viz['Mental_Health_Level'].map({0:'Poor',1:'Moderate',2:'Good'})

features = [
    'Sleep_Patterns',
    'Alcohol_Consumption',
    'Physical_Activity_Level',
    'History_of_Mental_Illness',
]

for col in features:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Mental_Health_Level', data=df_viz)

    plt.title(f"{col} vs Mental Health Level")
    plt.xticks(rotation=20)
    plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cols = [
    "Mental_Health_Level",
    "Physical_Activity_Level",
    "Alcohol_Consumption",
    "History_of_Mental_Illness",
    "Lifestyle_Risk"
]

corr = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")

plt.title("Correlation Heatmap (Key Features)")
plt.show()
```
